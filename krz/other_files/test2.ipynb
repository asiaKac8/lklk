```python
### libraries for model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
col_names = []
for (columnName, columnData) in df.iteritems():
    #print(columnName)
    col_names.append(columnName)

print(col_names)
```

```python
### classification

X = df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df['classification']

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.20, random_state=1)


# Model
model = SVC(gamma='auto') # SVC
model.fit(X_train, Y_train)

# Predykcja
predictions = model.predict(X_test)

# Pickle - zapis modelu, aby można go było użyć później
pd.to_pickle(model,r'.\\new_model.pickle')

model = pd.read_pickle(r'.\\new_model.pickle') 

# Tutaj użytkownik może wprowadzić dane z klawiatury
sepal_length = float(input("Enter sepal_length: "))
sepal_width = float(input("Enter sepa_width: "))
petal_length = float(input("Enter petal_length: "))
petal_width = float(input("Enter petal_width: "))

result = model.predict([[sepal_length,sepal_width,petal_length,petal_width]])
print(result)
```
